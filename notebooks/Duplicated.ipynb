```python
import pandas as pd

df = pd.read_csv(r"C:\Users\ibrahim.zimit\Desktop\Book3.csv")

duplicates = df[df.duplicated(subset=["ACCOUNT NUMBER"], keep=False)]

with pd.ExcelWriter(r"C:\Users\ibrahim.zimit\Desktop\output5.xlsx", engine="openpyxl") as writer:
    df.to_excel(writer, sheet_name="Original", index=False)
    duplicates.to_excel(writer, sheet_name="Duplicates", index=False)
```
