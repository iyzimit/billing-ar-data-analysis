```python
import pandas as pd

df_a = pd.read_excel(r"C:\Users\ibrahim.zimit\Desktop\A.xlsx")
df_b = pd.read_excel(r"C:\Users\ibrahim.zimit\Desktop\B.xlsx")

# Use first column regardless of name
a_accounts = df_a.iloc[:, 0].dropna().astype(str).str.strip().drop_duplicates()
b_accounts = df_b.iloc[:, 0].dropna().astype(str).str.strip().drop_duplicates()

# Rest stays the same
b_not_a = b_accounts[~b_accounts.isin(a_accounts)]
df_b_not_a = pd.DataFrame({"Account Number": b_not_a})

a_not_b = a_accounts[~a_accounts.isin(b_accounts)]
df_a_not_b = pd.DataFrame({"Account Number": a_not_b})

df_b_not_a.to_excel(r"C:\Users\ibrahim.zimit\Desktop\B_not_in_A.xlsx", index=False)
df_a_not_b.to_excel(r"C:\Users\ibrahim.zimit\Desktop\A_not_in_B.xlsx", index=False)

print(f"In B not in A: {len(df_b_not_a)}")
print(f"In A not in B: {len(df_a_not_b)}")
```
