```python
import pandas as pd
import glob
import os
from pathlib import Path
```

```python
# ================================
# 1. LOAD ALL FILES (SAFE)
# ================================

folder = Path(r"C:\Users\ibrahim.zimit\Desktop\2024")

files = [f for f in folder.glob("*.csv") if not f.name.startswith("~$")]

print("Folder exists:", folder.exists())

print("Valid files found:", len(files))

print("\n--- ROOT CONTENT ---")
for item in folder.iterdir():
    print(item.name)
```

```python
df_list = []

for file in files:
    try:
        print("Reading:", file)

        temp = pd.read_csv(
            file,
            low_memory=False,
            on_bad_lines="skip",
            encoding="latin1"
        )

        df_list.append(temp)

        print("✅ Loaded:", temp.shape)

    except Exception as e:
        print(f"❌ Failed {file}: {e}")

df = pd.concat(df_list, ignore_index=True)

print("Final shape:", df.shape)
```

```python
# ================================
# 6. AGGREGATE DATA
# ================================
summary = df.groupby(['state', 'billmonthname']).agg(
    customer_population=('accountno', 'nunique'),
    energy_billed=('consumption', 'sum'),
    amount_billed=('bill_amount', 'sum'),
    payment=('payment', 'sum')
).reset_index()
```

```python
# ================================
# 7. CREATE EXCEL OUTPUT
# ================================

output_file = r"C:\Users\ibrahim.zimit\Desktop\selected_accounts.xlsx"

with pd.ExcelWriter(output_file, engine='xlsxwriter') as writer:

    for state in summary['state'].unique():

        state_df = summary[summary['state'] == state]

        # Create Pivot Table
        pivot = state_df.pivot_table(
            index='month',
            values=[
                'customer_population',
                'energy_billed',
                'amount_billed',
                'payment'
            ],
            aggfunc={
                'customer_population': 'nunique',
                'energy_billed': 'sum',
                'amount_billed': 'sum',
                'payment': 'sum'
            }
        )

        # Sort months
        pivot = pivot.sort_index()

        # Transpose
        pivot = pivot.T

        # Rename rows
        pivot.index = [
            'Amount Billed (₦)',
            'Customer Population',
            'Energy Billed (kWh)',
            'Payment (₦)'
        ]

        # Fill missing values
        pivot = pivot.fillna(0)

        # Export to Excel
        pivot.to_excel(writer, sheet_name=state[:31])

print("✅ Report Generated Successfully!")
```
